**Importing Dependencies**

```python
# Getting the unique numbers of our training and test data
print(np.unique(Y_train))
print(np.unique(Y_test))
```

```python
# Scaling the values to a given range, i.e between 0 and 1
X_train = X_train/255
X_test = X_test/255
```

```python
print(X_train[10])
```

**Building a neural network**

```python
model = keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),
                          keras.layers.Dense(50, activation="relu"),
                          keras.layers.Dense(50, activation="relu"),
                          keras.layers.Dense(10, activation="sigmoid")
]
)
```

```python
# Compiling the neural network
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
# Training the neural network
model.fit(X_train, Y_train, epochs=10)
```

**Accuracy on Test data**

```python
loss, accuracy = model.evaluate(X_test, Y_test)
```

**Confusion Matrix**

```python
conf_mat = confusion_matrix(Y_test, Y_pred_labels)
```

```python
# Printing the confusion matrix
print(conf_mat)
```

```python
# Creating a heatmap to display the confusion matrix
plt.figure(figsize=(15, 7))
sns.heatmap(conf_mat, annot=True, square=True, cbar=True, fmt="d", cmap="Blues")
plt.ylabel("True labels")
plt.xlabel("Predicted labels")
plt.show()
```
